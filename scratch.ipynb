```python
import shutil
from pathlib import Path

# ---- Paths ----
src_root = Path("/Users/ayushjain/Development/Research/alphazero/runs/concept_vectors/BasicVectors")
dst_root = Path("/Users/ayushjain/Development/Research/alphazero/runs/novel/BasicVector")

# Ensure destination exists
dst_root.mkdir(parents=True, exist_ok=True)

novel_keys = [
    'pos_1449',
    'pos_2549',
    'pos_0038',
    'pos_2751',
    'pos_0001',
    'pos_2165',
    'pos_2361',
    'pos_2812',
    'pos_2027',
    'pos_3287',
    'pos_3134'
]

for key in novel_keys:
    src_path = src_root / key
    dst_path = dst_root / key
    
    if not src_path.exists():
        print(f"⚠️  Missing: {src_path}")
        continue
    
    if dst_path.exists():
        print(f"Skipping (already exists): {key}")
        continue
    
    print(f"Copying {key}...")
    shutil.copytree(src_path, dst_path)

print("Done.")
```
